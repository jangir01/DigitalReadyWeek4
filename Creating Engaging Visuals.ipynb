```python
x = np.array([1,2,3])
y = np.array([4,5,6]) ## Creating the coordinates

plt.scatter(x,y, color = 'Red') #Plots the points

x1 = np.array([1,2.5,3.1])
y1 = np.array([4.5,6,6])

plt.scatter(x1,y1, color = 'Blue') #Plots the points
plt.show() ## Shows the graph
```

## Pie Charts
Pie charts can be used to show percentages of a whole, and represents percentages at a set point in time. Unlike bar graphs and line graphs, pie charts do not show changes over time. The function to create a pie chart is 

`plt.pie()`

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

This pie chart will fill in the percentages that were given in the array. But notice that this pie chart isn't really giving us any information. We need to give it some meaning. 

We can add a labels **array** in the order of the percentages that each label is associated with. The syntax for this is

`plt.pie(y, labels=<labels array>)`


We can also add a **legend** with the line

`plt.legend(title = 'Title of Legend')`
