# Series and Data Frames
Two core components of pandas are Series and Data Frames. Series and Data Frames are two new **data structures** which we will use to analyze our data. A **Series** is essentially a column and a **Data Frame** is essentially a 2 dimensional table.


## Dictionaries 
A dictionary or a "dict" is python is a new kind of data type. It is similar to a list in that it stores multiple elements, however Dictionaries are used to store data values in key:value pairs. <br/>

Dictionaries are written with curly brackets, and have keys and values:

```python
thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)
```

The dictionary "thisdict" has three keys: brand, model, and year. The value for brand is "Ford", the value for "model" is "Mustang", and the value for "year", is 1964. <br/><br/>
